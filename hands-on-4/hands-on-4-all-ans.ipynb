# **Problem 1 with code and explanation**

```python
import heapq

def pradeep_merge_sorted_arrays(arrays):
    merged = []
    heap = [(array[0], i, 0) for i, array in enumerate(arrays) if array]
    heapq.heapify(heap)

    while heap:
        val, array_index, element_index = heapq.heappop(heap)
        merged.append(val)
        if element_index + 1 < len(arrays[array_index]):
            next_val = arrays[array_index][element_index + 1]
            heapq.heappush(heap, (next_val, array_index, element_index + 1))

    return merged

# Test cases
array1 = [1, 3, 5, 7]
array2 = [2, 4, 6, 8]
array3 = [0, 9, 10, 11]

print("Input:")
print("array1 =", array1)
print("array2 =", array2)
print("array3 =", array3)

output = pradeep_merge_sorted_arrays([array1, array2, array3])
print("\nOutput:")
print(output)
```

# **Problem 2 with code and explanation**

```python
def pradeep_remove_duplicates(array):
    if len(array) == 0:
        return []

    # Index to store the unique elements
    unique_index = 0

    # Traverse through the array starting from the second element
    for i in range(1, len(array)):
        # If current element is different from the previous element
        if array[i] != array[unique_index]:
            unique_index += 1
            array[unique_index] = array[i]

    # Slice the array to keep only unique elements
    return array[:unique_index + 1]

# Test cases
array1 = [2, 2, 2, 2, 2]
array2 = [1, 2, 2, 3, 4, 4, 4, 5, 5]

print("Input:")
print("array1 =", array1)
print("array2 =", array2)

output1 = pradeep_remove_duplicates(array1)
output2 = pradeep_remove_duplicates(array2)

print("\nOutput:")
print("array1 =", output1)
print("array2 =", output2)
```
