```python
import timeit
import matplotlib.pyplot as plt
import platform
import psutil

#for printing processor and ram info
def print_system_info():
    print("Processor:", platform.processor())
    print("RAM:", psutil.virtual_memory().total / (1024 ** 3), "GB")

def custom_bubble_sort(arr):
    n = len(arr)
    # Traverse through all array elements
    for i in range(n):
        swapped = False
        # Last i elements are already in place
        for j in range(0, n-i-1):
            # Swap if the element found is greater than the next element
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                swapped = True
        # If no elements were swapped in the inner loop, then the array is sorted
        if not swapped:
            break
    return arr

def benchmark_custom_bubble_sort():
    input_sizes = [5, 10, 20, 50, 100,500, 1000,5000]
    execution_times = []

    for size in input_sizes:
        array = list(range(size, 0, -1))  # Descending order list for worst-case scenario
        time_taken = timeit.timeit(lambda: custom_bubble_sort(array.copy()), number=10)
        execution_times.append(time_taken)

    # Plotting the results
    plt.figure(figsize=(10, 6))
    plt.plot(input_sizes, execution_times, label='Bubble Sort', marker='o', color='red')
    plt.xlabel('Input Size (n)')
    plt.ylabel('Execution Time (seconds)')
    plt.title('Custom Bubble Sort Benchmark')
    plt.grid(True)
    plt.legend()
    plt.show()

benchmark_custom_bubble_sort()
```
