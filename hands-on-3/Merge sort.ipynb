```python
# Merge function to merge two halves
def merge(left, right):
    result = []
    i = j = 0

    # Traverse both lists and insert the smaller element into result
    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    # Append remaining elements of the left and right parts, if any
    result.extend(left[i:])
    result.extend(right[j:])

    return result

# Merge sort function to divide the array and merge them back
def custom_merge_sort(array):
    if len(array) <= 1:
        return array

    # Divide the array into two halves
    mid = len(array) // 2
    left_half = custom_merge_sort(array[:mid])
    right_half = custom_merge_sort(array[mid:])

    # Merge the sorted halves
    return merge(left_half, right_half)

# This function takes input from the user
get_user_input = input("Please enter the array elements, separated by commas: ")
# This will convert input string to integers
array = list(map(int, get_user_input.split(',')))

# Function is used to sort the array using merge sort
final_sorted_array = custom_merge_sort(array)

# This function will print the sorted array
print("Final sorted array:", final_sorted_array)
```
