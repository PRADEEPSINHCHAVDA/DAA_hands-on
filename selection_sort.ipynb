```python
import timeit
import matplotlib.pyplot as plt
import platform
import psutil

#for printing processor and ram info
def print_system_info():
    print("Processor:", platform.processor())
    print("RAM:", psutil.virtual_memory().total / (1024 ** 3), "GB")

print_system_info()

def custom_selection_sort(arr):
    # Traverse through all array elements
    for i in range(len(arr)):
        min_index = i
        # Find the minimum element in remaining unsorted array
        for j in range(i + 1, len(arr)):
            if arr[j] < arr[min_index]:
                min_index = j
        # Swap the found minimum element with the first element
        arr[i], arr[min_index] = arr[min_index], arr[i]
    return arr

def benchmark_custom_selection_sort():
    input_sizes = [5, 10, 20, 50, 100,500, 1000,5000]
    execution_times = []

    for size in input_sizes:
        array = list(range(size, 0, -1))  # Descending order list for worst-case scenario
        time_taken = timeit.timeit(lambda: custom_selection_sort(array.copy()), number=10)
        execution_times.append(time_taken)

    # Plotting the results
    plt.figure(figsize=(10, 6))
    plt.plot(input_sizes, execution_times, label='Selection Sort', marker='o', color='green')
    plt.xlabel('Input Size (n)')
    plt.ylabel('Execution Time (seconds)')
    plt.title('Custom Selection Sort Benchmark')
    plt.grid(True)
    plt.legend()
    plt.show()

benchmark_custom_selection_sort()
```
