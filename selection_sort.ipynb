this file contains the entire code of sorting through selection sort and the code of benchmarking use to plot graph.

```python
import timeit
import matplotlib.pyplot as plt

def demo_selection_sort(array):
    # Iterate till end of the elements
    for i in range(len(array)):
        # to find minimum element in array,assume the first index as minimum
        minimum_index = i
        #find the index of minimum element in array
        for j in range(i + 1, len(array)):
            if array[j] < array[minimum_index]:
                minimum_index = j
        # if minimum array found then swap it with the first element
        array[i], array[minimum_index] = array[minimum_index], array[i]

    return array

# this function takes input from the user
get_user_input = input("please enter the array elements, separated by commas: ")
#this will convert input string to integers
array = list(map(int, get_user_input.split(',')))

# function is use to Sort the array
final_sorted_array = demo_selection_sort(array)

# this function will Print the sorted array
print("Final sorted array:", final_sorted_array)

def benchmark_demo_selection_sort():
    input_arraysizes = [5, 10, 50, 100,500,5000,10000]#list of different list to sort
    execution_times = []

    for size in input_arraysizes:
        array = list(range(size, 0, -1))  # Descending order list for worst-case scenario
        time_taken = timeit.timeit(lambda: demo_selection_sort(array.copy()), number=10)#measuring the time taken to sort
        execution_times.append(time_taken)

    # below code will plot the results
    plt.figure(figsize=(10, 6))
    plt.plot(input_arraysizes, execution_times, label='Selection Sort', marker='o', color='green')
    plt.xlabel('Input Size (n)')
    plt.ylabel('Execution Time (seconds)')
    plt.title('Pradeep_1002231638 Selection Sort Benchmark')
    plt.grid(True)
    plt.legend()
    plt.show()
#calling the benchmark function
benchmark_demo_selection_sort()
```
