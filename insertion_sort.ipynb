```python
import platform
import psutil

def print_system_info():
    print("Processor:", platform.processor())
    print("RAM:", psutil.virtual_memory().total / (1024 ** 3), "GB")
    print("Python version:", platform.python_version())

print_system_info()

import timeit
import matplotlib.pyplot as plt

def custom_insertion_sort(arr):
    # Traverse till end of the array
    for i in range(1, len(arr)):
        key_item = arr[i]
        # Move elements of array[0..i-1], that are greater than key_item, to one position ahead of their current position
        j = i - 1
        while j >= 0 and arr[j] > key_item:
            arr[j + 1] = arr[j]
            j -= 1
        # Place the key item at after the last smaller item
        arr[j + 1] = key_item
    return arr

def benchmark_custom_insertion_sort():
    input_sizes = [5, 10, 20, 50, 100, 500, 1000, 5000]
    execution_times = []

    for size in input_sizes:
        array = list(range(size, 0, -1))  # Descending order list for worst-case scenario
        time_taken = timeit.timeit(lambda: custom_insertion_sort(array.copy()), number=10)
        execution_times.append(time_taken)

    # Plotting the results in graph
    plt.figure(figsize=(10, 6))
    plt.plot(input_sizes, execution_times, label='Insertion Sort', marker='o', color='blue')
    plt.xlabel('Input Size (n)')
    plt.ylabel('Execution Time (seconds)')
    plt.title('Custom Insertion Sort Benchmark')
    plt.grid(True)
    plt.legend()
    plt.show()

benchmark_custom_insertion_sort()

```
