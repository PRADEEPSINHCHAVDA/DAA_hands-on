this is the demo of insertion sort and benchmark plot of the values


```python
import timeit
import matplotlib.pyplot as plt

def demo_insertion_sort(array):
    # Traverse through 1 to len(array)
    for i in range(1, len(array)):
        key = array[i]
        j = i - 1
        # Move elements of array[0..i-1], that are greater than key, to one position ahead of their current position
        while j >= 0 and array[j] > key:
            array[j + 1] = array[j]
            j -= 1
        array[j + 1] = key

    return array

# This function takes input from the user
get_user_input = input("Please enter the array elements, separated by commas: ")
# This will convert input string to integers
array = list(map(int, get_user_input.split(',')))

# Function is used to sort the array
final_sorted_array = demo_insertion_sort(array)

# This function will print the sorted array
print("Final sorted array:", final_sorted_array)

def benchmark_demo_insertion_sort():
    input_arraysizes = [5, 10, 50, 100,500,5000,10000]#list of different list to sort
    execution_times = []

    for size in input_arraysizes:
        array = list(range(size, 0, -1))  # Descending order list for worst-case scenario
        time_taken = timeit.timeit(lambda: demo_insertion_sort(array.copy()), number=10)#measuring the time taken to sort
        execution_times.append(time_taken)

    # Plotting the results in graph
    plt.figure(figsize=(10, 6))
    plt.plot(input_arraysizes, execution_times, label='Insertion Sort', marker='o', color='blue')
    plt.xlabel('Input Size (n)')
    plt.ylabel('Execution Time (seconds)')
    plt.title('Pradeep_1002231638 Selection Sort Benchmark')
    plt.grid(True)
    plt.legend()
    plt.show()

#calling the benchmark function
benchmark_demo_insertion_sort()
```
